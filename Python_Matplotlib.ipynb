```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["mat", "chem", "clang", "eng"])
y = np.array([10, 9, 5, 9])

plt.bar(x,y, width = 0.1)
plt.barh(x, y,color = "red", height = 0.1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=  np.array(['A','B','C','D'])
y = np.array([30, 30, 25, 15])
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "hotpink", "b", "#4CAF50"]


plt.pie(y,labels=x,explode = myexplode,shadow=True,colors=mycolors)
plt.title("pie chart")
plt.legend(title="moon")
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 6, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([12, 8, 6, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```
