```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6, 23, 567, 789])
ypoints = np.array([0, 5, 45, 56, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["mat", "chem", "clang", "eng"])
y = np.array([10, 9, 5, 9])

plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```
